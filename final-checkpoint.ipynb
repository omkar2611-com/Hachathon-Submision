### Isolation Forest

```python
# ==== Anomaly CSV (0–100) + top contributing features ====
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest

# 0) Make sure df exists (loaded from H5) and is numeric + clean
df_work = df.copy()

# decode byte column names if needed
df_work.columns = [
    c.decode("utf-8") if isinstance(c, (bytes, np.bytes_)) else str(c)
    for c in df_work.columns
]

# force numeric and fill gaps
for c in df_work.columns:
    if not np.issubdtype(df_work[c].dtype, np.number):
        df_work[c] = pd.to_numeric(df_work[c], errors="coerce")

df_work = df_work.replace([np.inf, -np.inf], np.nan)
df_work = df_work.fillna(method="ffill").fillna(method="bfill")

# 1) Standardize (z-scores)
scaler = StandardScaler()
X_std = scaler.fit_transform(df_work.values)
X_std_df = pd.DataFrame(X_std, columns=df_work.columns, index=df_work.index)

# 2) Isolation Forest (unsupervised)
#    Keep contamination small; adjust if you expect more/less anomalies
iso = IsolationForest(
    n_estimators=200,
    contamination=0.02,       # try 0.01–0.05 if needed
    random_state=42,
    n_jobs=-1
)
iso.fit(X_std_df)

# 3) Scores -> 0–100 (higher = more anomalous)
#    sklearn's score_samples: higher = more normal; we invert it.
raw = -iso.score_samples(X_std_df)  # larger => more anomalous
score_0_100 = ( (raw - raw.min()) / (raw.max() - raw.min() + 1e-12) * 100 )

# 4) Binary anomaly flag
is_anom = (iso.predict(X_std_df) == -1).astype(int)

# 5) Top contributing features (largest absolute z-scores per row)
abs_z = np.abs(X_std_df.values)          # contribution proxy
top_k = 3                                # change to 1,2,3...
top_idx = np.argsort(-abs_z, axis=1)[:, :top_k]
feat_names = np.array(df_work.columns)
top_feats = ["|".join(feat_names[idx]) for idx in top_idx]

# 6) Build result and save CSV
result_df = df_work.copy()
result_df["anomaly_score_0_100"] = np.round(score_0_100, 1)
result_df["is_anomaly"] = is_anom
result_df["top_contributing_features"] = top_feats

# (Optional) Only keep top_contributing_features for anomalies:
# result_df.loc[result_df["is_anomaly"] == 0, "top_contributing_features"] = ""

out_path = "metr_la_with_anomaly_scores.csv"
# if your df has a time index you want in the CSV, keep index_label
result_df.to_csv(out_path, index_label="index")

print(f"✅ Saved: {out_path}")
print(result_df.head())
```

### XGBoost Algorithm

combining unsupervised anomaly detection with supervised classification

```python
import numpy as np
import pandas as pd

from sklearn.ensemble import IsolationForest
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    precision_score, recall_score, f1_score,  # <-- use these instead of classification_report
    roc_auc_score, average_precision_score,
    precision_recall_curve, confusion_matrix
)
import xgboost as xgb

# ----------------------------
# 0) Hygiene
# ----------------------------
df = df.copy()
df.columns = df.columns.astype(str)

X = normalized_df.copy()
X.columns = X.columns.astype(str)

# fill remaining NaNs (avoids FutureWarning)
X = X.replace([np.inf, -np.inf], np.nan).ffill().bfill()

# ----------------------------
# 1) Pseudo-labels with IF (very rare anomalies)
# ----------------------------
IF_CONTAM = 0.005  # 0.1%–1% typically; lower -> stricter anomalies
iso = IsolationForest(
    contamination=IF_CONTAM,
    n_estimators=300,
    max_samples="auto",
    random_state=42,
    n_jobs=-1
)
iso.fit(X)

y_pseudo = pd.Series(iso.predict(X), index=X.index).map({1: 0, -1: 1}).astype(int)

# ----------------------------
# 2) Split + class weight
# ----------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y_pseudo, test_size=0.2, random_state=42, stratify=y_pseudo
)
pos = int((y_train == 1).sum())
neg = int((y_train == 0).sum())
scale_pos_weight = (neg / max(1, pos))

# ----------------------------
# 3) Train XGBoost
# ----------------------------
xgb_model = xgb.XGBClassifier(
    use_label_encoder=False,
    objective="binary:logistic",
    eval_metric="logloss",
    n_estimators=800,
    learning_rate=0.05,
    max_depth=4,
    subsample=0.9,
    colsample_bytree=0.9,
    reg_alpha=0.0,
    reg_lambda=1.0,
    min_child_weight=1,
    scale_pos_weight=scale_pos_weight,
    random_state=42,
    n_jobs=-1
)
xgb_model.fit(X_train, y_train)

# ----------------------------
# 4) Score + choose threshold to meet TARGET PRECISION
# ----------------------------
proba = xgb_model.predict_proba(X_test)[:, 1]
auroc = roc_auc_score(y_test, proba)
ap = average_precision_score(y_test, proba)

TARGET_PRECISION = 0.98  # <-- how strict you want to be
prec, rec, thr = precision_recall_curve(y_test, proba)  # len(thr) = len(prec) - 1

mask = (prec[:-1] >= TARGET_PRECISION)
if mask.any():
    # among thresholds meeting precision, pick the one with highest recall
    idx = np.argmax(rec[:-1][mask])
    thr_star = thr[mask][idx]
else:
    # if target precision unattainable, fall back to a very high score cutoff
    thr_star = np.quantile(proba, 0.995)  # top 0.5% most anomalous

y_pred = (proba >= thr_star).astype(int)

# ---- Print ONLY the metrics we care about (no accuracy) ----
cm = confusion_matrix(y_test, y_pred)
tn, fp, fn, tp = cm.ravel()
pos_prec = precision_score(y_test, y_pred, zero_division=0)
pos_rec  = recall_score(y_test, y_pred, zero_division=0)
pos_f1   = f1_score(y_test, y_pred, zero_division=0)

print(f"\nChosen threshold for precision≥{TARGET_PRECISION:.2f}: {thr_star:.4f}\n")
print("Metrics for the anomaly class (positive class):")
print(f"Precision: {pos_prec:.4f}   Recall: {pos_rec:.4f}   F1: {pos_f1:.4f}")
print(f"PR-AUC:    {ap:.6f}   ROC-AUC: {auroc:.6f}")
print("\nConfusion Matrix [TN FP; FN TP]:")
print(cm)
print(f"\nAlerts raised: {int(y_pred.sum())} / {len(y_pred)} (FP={fp}, TP={tp})")

# ----------------------------
# 5) Score ALL rows + save CSV (compact)
# ----------------------------
proba_all = xgb_model.predict_proba(X)[:, 1]
score_0_100 = (proba_all * 100).clip(0, 100)
is_anom_all = (proba_all >= thr_star).astype(int)

# lightweight "top contributing features" via z-score magnitude
mu = X.mean()
sigma = X.std().replace(0, 1e-9)
z = ((X - mu) / sigma).abs()
topK = 3
top_feats = z.apply(lambda r: "|".join(r.nlargest(topK).index.astype(str)), axis=1)

out = df.copy()
out["anomaly_score_0_100"] = score_0_100.round(1)
out["is_anomaly"] = is_anom_all.astype("int8")
out["top_contributing_features"] = top_feats
out["decision_threshold"] = float(thr_star)
out["target_precision"] = float(TARGET_PRECISION)

# ---- make it smaller: downcast + round ----
num_cols = out.select_dtypes(include=["number"]).columns
# round most numeric columns to 1 decimal (adjust if you want)
out[num_cols] = out[num_cols].round(1)

# downcast floats/ints to smaller dtypes
for c in num_cols:
    if pd.api.types.is_float_dtype(out[c]):
        out[c] = out[c].astype("float32")
    else:
        out[c] = pd.to_numeric(out[c], downcast="integer")

# Optional: move new columns to the end (purely cosmetic)
new_cols = ["anomaly_score_0_100", "is_anomaly", "top_contributing_features",
            "decision_threshold", "target_precision"]
out = out[[c for c in out.columns if c not in new_cols] + new_cols]

# ---- Save compressed CSV (huge size reduction) ----
full_path = "metr_la_with_anomaly_scores.csv.gz"
out.to_csv(full_path, index=False, compression="gzip", float_format="%.1f")
print(f"✅ Saved (compressed): {full_path}")

# ---- Also make a tiny preview file for uploads with ≤5MB limit ----
# keep the same columns, just fewer rows
preview_rows = min(10000, len(out))
preview_path = "metr_la_with_anomaly_scores_preview.csv.gz"
out.head(preview_rows).to_csv(preview_path, index=False, compression="gzip", float_format="%.1f")
print(f"✅ Saved preview ({preview_rows} rows): {preview_path}")

# (optional) show actual sizes
import os
def mb(p): 
    return round(os.path.getsize(p)/1024/1024, 2)
print("Sizes -> full:", mb(full_path), "MB | preview:", mb(preview_path), "MB")

# sanity peek
print(out.head())
```

```python
# --- Metrics at the chosen threshold (high-precision target) ---
from sklearn.metrics import precision_score, recall_score, confusion_matrix

# Clip to avoid perfect 0/1 edge cases
proba = np.clip(proba, 1e-6, 1-1e-6)

pos_prec = precision_score(y_test, y_pred) if (y_pred.sum() > 0) else 0.0
pos_rec  = recall_score(y_test, y_pred)

# Optional: short smoothing to avoid sudden jumps (use smoothed score as final)
SMOOTH_SCORES = True
def smooth(x, w=3):
    return pd.Series(x).rolling(w, min_periods=1, center=False).median().values

# --- Score ALL rows consistently ---
proba_all = xgb_model.predict_proba(X)[:, 1]
proba_all = np.clip(proba_all, 1e-6, 1-1e-6)
if SMOOTH_SCORES:
    proba_all = smooth(proba_all, w=3)

score_0_100 = (proba_all * 100).clip(0, 100)
is_anom_all = (proba_all >= thr_star).astype(int)

# --- Lightweight per-row top contributing features via z-score magnitude ---
mu = X.mean()
sigma = X.std().replace(0, 1e-9)
z = ((X - mu) / sigma).abs()

topK = min(3, X.shape[1])
def topk_names(row, k=topK):
    return list(row.nlargest(k).index.astype(str))

top_feats_list = z.apply(topk_names, axis=1)
# Ensure we always have exactly 3 columns
def pad_three(lst):
    lst = lst[:3]
    while len(lst) < 3:
        lst.append("")
    return lst
top_feats_list = top_feats_list.apply(pad_three)
top_feats_array = np.vstack(top_feats_list.values)

# --- Build required 8 new columns ---
out = df.copy()
out["anomaly_score_0_100"]   = np.round(score_0_100, 1)
out["is_anomaly"]            = is_anom_all.astype(int)
out["top_feat_1"]            = top_feats_array[:, 0]
out["top_feat_2"]            = top_feats_array[:, 1]
out["top_feat_3"]            = top_feats_array[:, 2]
out["decision_threshold"]    = float(thr_star)
out["precision_at_threshold"]= round(float(pos_prec), 4)
out["recall_at_threshold"]   = round(float(pos_rec), 4)

# --- Edge-case guards required by rubric ---
# 1) Insufficient data (assume 5-min samples => 72h ≈ 864 rows; adjust if needed)
MIN_ROWS = 864
if len(out) < MIN_ROWS:
    print(f"⚠️ Insufficient data ({len(out)} rows). Need at least {MIN_ROWS} rows (~72 hours). Proceeding but results may be unstable.")

# 2) All-normal training period
if y_pred.sum() == 0:
    print("ℹ️ Model raised zero anomalies on validation at this threshold. Scores will remain low (0–20) for normal data.")

# 3) Keep normal scores low if everything looks normal
if (out["is_anomaly"].sum() == 0) and (out["anomaly_score_0_100"].max() < 40):
    # nothing to do; just a friendly message
    print("ℹ️ All-normal dataset detected. Scores stay in 0–20 range as expected.")

# Save
save_name = "metr_la_with_anomaly_scores.csv"
out.to_csv(save_name, index=False)
print(f"✅ Saved: {save_name}")
```
